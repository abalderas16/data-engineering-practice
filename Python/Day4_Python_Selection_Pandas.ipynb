```python
#*****PRODUCTION STYLE*****
import pandas as pd

df = pd.read_csv("Contr_02_17_V2.csv")

less_mileage_orders = df.loc[df["Miles"] <= 30].copy() #Make a copy of the data

less_mileage_orders # Filter orders with Miles less than or equal to 30
```

```python
#*****ENGINEERING STYLE*****
import pandas as pd

def less_than_30(df, threshold=30):
  return df.loc[df["Miles"] <= threshold].copy()

low_mileage_orders = less_than_30(df)

low_mileage_orders # Filter orders with Miles less than or equal to 30
```

**# Complex Miles and Delay Filter**

```python
#*****PRODUCTION STYLE*****
#Filter by Low Miles and Delay #Note Delay is an Object type of data

import pandas as pd

df = pd.read_csv("Contr_02_17_V2.csv")

#Removes leading & trailing spaces from column names "Delay " to "Delay"
df.columns = df.columns.str.strip()

#Converts Miles to numeric safely, errors="coerce" converts it to NaN istead of crashing for bad values
df["Miles"] = pd.to_numeric(df["Miles"], errors="coerce")

# Custom function to convert 'HH:MM' string to timedelta
def parse_delay_to_timedelta(delay_str):
    if pd.isna(delay_str): #This checks if the vaule is missing if so returns pd.NaT
        return pd.NaT #NaT stands for Not A Time

    parts = str(delay_str).split(':') #This becomes from this "0:30" to ["0","30"]
    if len(parts) == 2: #Make sure it has twp exactly 2 parts ("HH:MM")
        try:
            hours = int(parts[0]) #Convert hours into Integers (hours = 0)
            minutes = int(parts[1]) #Conver minutes into Integers
            return pd.Timedelta(hours=hours, minutes=minutes)#Now "0.30" becomes 0 days 00:30:00
        except ValueError:
            return pd.NaT #If fails return NaT

    return pd.NaT#This function ALWAYS returns either a Timedelta or NaT

# Apply function to the entire Delay Column
df["Delay_td"] = df["Delay"].apply(parse_delay_to_timedelta)
#df["Delay_td"]creates new column, Runs All custom function on every row in Delay Column .apply(parse_delay_to_timedelta)

# Filter: Filtering using .loc with Multiple conditions Miles <= 30 AND Delay <= 30 minutes
filtered = df.loc[
    (df["Miles"].notna()) & # Keeps only rows where Miles is NOT missing (Nan)
    (df["Miles"] <= 30) & #Keeps only rows where Miles is <= 30
    (df["Delay_td"].notna()) & # Keeps only rows where Delay was successfully prased not (NaT)
    (df["Delay_td"] <= pd.Timedelta(minutes=30)) #Keeps only rows with Delay <= 30 mintues
].copy() #Copy makes filtered an independent DataFrame

filtered [["Cust", "Order Number", "Miles", "Delay", "Delay_td"]]
```
